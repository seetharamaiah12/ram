```
def emp_data():
  p=input("Employee Name : ")
  q=input("Employee Number : ")
  r=input("Week Ending :  ")
  s=float(input("Number Of Hours Worked : "))
  t=float(input("Hourly Rate : "))
  u=float(input("Overtime Rate : "))
  v=float(input("Standard Tax Rate : "))
  w=float(input("Overtime Tax Rate :"))
 
  if s>37.5:
     hours_overtime = s-37.5
  else:
      hours_overtime =0  
  print("hours_overtime : {}".format(hours_overtime))
  
  hours_normal = s-hours_overtime
  e1 =hours_normal* t
  d1 = v/100*e1
  
  e2 =hours_overtime*(t*u)
  d2=w/100*e2
  
  total_pay = e1+e2
  total_detections =d1+d2
  net_pay = total_pay - total_detections
  print( "\t\t\t\t\t\t\t PAYSLIP")
  print("WEEK ENDING : {} \n Employee : {} \n Employee Number : {}".format(r,p,q))
  print("\t\t\t Earnings \t\tDeductions")
  print("\t\t\t Hours\t Rate \t Total")
  print("Hours (normal)\t\t  {}\t{}\t{}\t Tax @ {}%\t {}".format(hours_normal,t,e1,v,d1))
  print("Hours (overtime)\t  {}\t{}\t{}\t Tax @ {}%\t {}".format(hours_overtime,t*u,e2,w,d2))
  print("\t\t\t Total Pay :\t\t\t\t {}\n\t\t\t Total Detections : \t\t\t {} \n\t\t\t Net Pay :\t\t\t\t{}".format(total_pay,total_detections, net_pay))
  
  
  





```
